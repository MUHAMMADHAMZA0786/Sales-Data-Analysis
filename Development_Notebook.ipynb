function declartion: total_revenue_profit_by_sales_channel(sales_data)

    group sales_data by 'Sales_Channel' and calculate the sum of 'Total_Revenue' and 'Total_Profit' for each channel, store it in total_revenue_profit_by_sales_channel

    get the unique sales channels from total_revenue_profit_by_sales_channel

    set the width of each bar as bar_width

    generate positions for the bars using np.arange and the length of sales_channels, store it in positions

    plot a bar graph with positions as the x-axis, total_revenue_profit_by_sales_channel['Total_Revenue'] as the y-axis, and width=bar_width, labeled as 'Total Revenue'

    plot another bar graph with positions + bar_width as the x-axis, total_revenue_profit_by_sales_channel['Total_Profit'] as the y-axis, and width=bar_width, labeled as 'Total Profit'

    set the x-axis ticks and labels as positions + bar_width / 2 and sales_channels

    set the x-axis label as 'Sales Channel'
    set the y-axis label as 'Amount'
    set the title of the graph as 'Total Revenue and Profit by Sales Channel'
    set the legend

    show the plot
